```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dyn(p, c, g):
    # A discrete logistic model (Verhulst dynamics)
    # c = carrying capacity
    # g = growth rate.  Chaos transition between g=2 and g=3.
    new_p = p + p*(c-p)*g
    if new_p < 0:
        new_p = 0
    return new_p
```

```python
def plotpop(start, time, capacity, growth):
    population = [start]
    for t in range(1,time):
        population += [dyn(population[t-1], capacity, growth)]
    c = [capacity]*time
    plt.plot(range(time), population)
    plt.plot(range(time),c)
    plt.title('Population Dynamics')
    plt.xlabel('Time')
    plt.ylabel('Population')
    plt.show()
    return ()
```

```python
def predprey(x,y,alpha,beta,gamma,delta):
    # A discrete Lotka-Volterra model.
    # x is the prey population.
    # y is the predator population.
    new_x = x + x*(alpha - beta*y)
    new_y = y - y*(gamma - delta*x)
    if new_x < 0:
        new_x = 0
    if new_y < 0:
        new_y = 0
    return new_x,new_y
```

```python
def plotpp(startx, starty, time, a,b,c,d):
    # Makes a plot with x-axis prey
    # and y-axis predators.  
    px = [startx]
    py = [starty]
    for t in range(1,time):
        x,y = px[t-1], py[t-1]
        nx,ny = predprey(x,y,a,b,c,d)
        px += [nx]
        py += [ny]
    plt.plot(px, py)
    plt.title('Population Dynamics')
    plt.xlabel('Prey population')
    plt.ylabel('Predator population')
    plt.show()
    return ()
```

```python
def plotppt(startx, starty, time, a,b,c,d):
    # Makes a plot with time on x-axis.
    # Two curves, one for predators, one for prey.
    px = [startx]
    py = [starty]
    for t in range(1,time):
        x,y = px[t-1], py[t-1]
        nx,ny = predprey(x,y,a,b,c,d)
        px += [nx]
        py += [ny]
    plt.plot(range(time), px, label='Prey')
    plt.plot(range(time), py, label='Predator')
    plt.title('Population Dynamics')
    plt.xlabel('Time')
    plt.legend()
    plt.show()
    return ()
```

```python
%pylab inline
plotppt(12,10,300,0.1,0.01,0.1,0.01)
plotpp(12,10,300,0.1,0.01,0.1,0.01)
```
